```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath = "/Users/Rithik/Desktop/Desktop - Rithik’s MacBook Pro/Desktop/Folders/Year2/Semester 1/DS 4200/Final Project/Airline_Reviews.csv"
df = pd.read_csv(filepath)
df_data = df.drop(columns = ['Route', 'Inflight Entertainment', 'Aircraft', 'Wifi & Connectivity', 'Recommended', 'Review Date',])
df_data.head(20)
```

```python
# Convert all values in the collumn 'Overall_Rating' to numeric vlaues
# Any non-numeric numbers changed to NaN
df_data['Overall_Rating'] = pd.to_numeric(df_data['Overall_Rating'], errors='coerce')

# Calculate the mean
means_rating = df_data.groupby("Airline Name")['Overall_Rating'].mean().astype(int)
df_data = df_data.dropna(subset=['Overall_Rating']) # Drop all NaN Values
```

```python
# Only use the top 25 airlines
means_rating_top25 = means_rating.sort_values(ascending=False).head(25)

# Create a bar graph for the top 25 sorted means_rating
plt.figure(figsize=(10, 6))
means_rating_top25.plot(kind='bar')

# Add title and labels
plt.title('Top 20 Airlines by Average Overall Rating', fontsize=14)
plt.xlabel('Airline Name', fontsize=12)
plt.ylabel('Mean Overall Rating', fontsize=12)

# Display the bar graph
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
