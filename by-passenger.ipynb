```python
import pandas as pd

with open ('Airline_Reviews.csv', 'r', encoding='utf-8') as infile:
    
    df = pd.read_csv(infile)
    
df
```

```python
columns = ['Type Of Traveller', 
           'Seat Comfort', 'Cabin Staff Service', 'Food & Beverages', 'Ground Service', 'Inflight Entertainment', 'Wifi & Connectivity', 'Value For Money']

df_travel = df.loc[:, columns].groupby('Type Of Traveller').mean()
df_travel.index[0]
```

```python
df_stacked = {'Type Of Traveller': [],
              'Category': [],
              'Rating': []}

for column in columns[1:]:
    for i in range(len(df_travel)):
        df_stacked['Type Of Traveller'].append(df_travel.index[i])
        df_stacked['Category'].append(column)
        df_stacked['Rating'].append(df_travel.loc[df_travel.index[i], column])
        
df_stacked = pd.DataFrame(df_stacked)
df_stacked
```

```python
import altair as alt

travel_colors = alt.Scale(domain=list(df_travel.index),
                          range=['black', 'red', 'green', 'blue'])

alt.Chart(df_stacked).mark_bar().encode(
    x=alt.X('Type Of Traveller:N'),
    y=alt.Y('Rating:Q', scale=alt.Scale(domain=[0, 5])),
    color=alt.Color('Type Of Traveller:N', scale=travel_colors),
    column='Category:N' 
).properties(
    title=alt.Title('Average Ratings for each Traveller Type', subtitle='o')
).configure_title(
    orient='top',
    anchor='middle',
    font='Times New Roman',
    fontSize=20,
    subtitleColor='white',
    subtitlePadding=5
).configure_legend(
    orient='none',
    legendX=290,
    legendY=10,
    fillColor='#EEEEEE',
    cornerRadius=10,
    padding=10,
    strokeColor='black'
).configure_header(
    title=None,
    labelFontSize=11,
    labelFontStyle='bold'
).configure_axisX(
    labels=False,
    ticks=False,
    title=None
).configure_axisY(
    labelFontSize=14,
    tickCount=5
)
```

```python
base = alt.Chart(df_stacked).mark_bar().encode(
       x=alt.X('Type Of Traveller:N'),
       y=alt.Y('Rating:Q', scale=alt.Scale(domain=[0, 5]), title='Average Rating'),)

bars = base.mark_bar().encode(
    color=alt.Color('Type Of Traveller:N', scale=travel_colors),
    tooltip=[
        alt.Tooltip('Type Of Traveller:N'),
        alt.Tooltip('Rating:Q', format='.2f'),
    ]
    )

text = base.mark_text(color='white', dx=19, dy=1, angle=90, fontStyle='bold').encode(
    text=alt.Text('Rating:Q', format=',.2f')
)

facet = alt.layer(bars, text, data=df_stacked).facet(
    column='Category:N'
)
facet
```

```python
final = facet.properties(
    title=alt.Title('Average Ratings for each Traveller Type')
).configure_title(
    orient='top',
    anchor='middle',
    font='Times New Roman',
    fontSize=20,
    dy=10
).configure_legend(
    orient='none',
    legendX=292,
    legendY=10,
    fillColor='#EEEEEE',
    cornerRadius=10,
    padding=10,
    strokeColor='black'
).configure_header(
    labelFontSize=11,
    labelFontStyle='bold',
    labelPadding=-320,
    titleOrient='bottom',
    titleFontSize=14,
).configure_axisX(
    labels=False,
    ticks=False,
    title=None
).configure_axisY(
    labelFontSize=13,
    tickCount=5,
    titleFontSize=15
)
final
```

```python
final.save('rating.html')
```
