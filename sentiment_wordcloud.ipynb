```python
import pandas as pd

with open ('nlp2.csv', 'r', encoding='utf-8') as infile:
    
    df = pd.read_csv(infile)
    
df
```

```python
from ast import literal_eval

df['Pos-Neg'] = ''

for i in range(len(df['Sentiment'])):
    compound = literal_eval(df.loc[i, 'Sentiment'])['compound']
    
    if compound > 0:
        df.loc[i, 'Pos-Neg'] = 'Positive'
    else:
        df.loc[i, 'Pos-Neg'] = 'Negative'
    
df['Pos-Neg']
```

```python
df
```

```python
cloud_dict = {'Airline Name': [],
              'Word': [],
              'Pos-Neg': []}

for i in range(len(df)):
    
    tokens = literal_eval(df.loc[i, 'Tokens'])
    
    for j in tokens:
        cloud_dict['Airline Name'].append(df.loc[i, 'Airline Name'])
        cloud_dict['Word'].append(j)
        cloud_dict['Pos-Neg'].append(df.loc[i, 'Pos-Neg'])

print(cloud_dict)
```

```python
import string

df_cloud = pd.DataFrame(cloud_dict).loc[:, ['Word', 'Pos-Neg']]

def punk(text):
    return all(x in string.ascii_lowercase for x in text)

df_cloud = df_cloud.loc[df_cloud['Word'].apply(lambda x: punk(x))]
df_cloud['Sum'] = 1
df_cloud = df_cloud.groupby(['Word', 'Pos-Neg']).count().reset_index()
df_cloud = df_cloud.loc[df_cloud['Sum'] > 1000]
```

```python
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
stopwords = ['airline', 'flight', 'airport', 'even', 'also', 'air', 'airlines', 'asked', 'back', 'another',
             'class', 'could', 'day', 'days', 'due', 'ever', 'flights', 'fly', 'flying', 'gate', 'get', 'go',
             'got', 'however', 'made', 'make', 'much', 'never', 'next', 'one', 'people', 'pm', 'quite', 'really', 'said', 'still', 'take', 'though', 'told', 'took', 'travel', 'trip', 'two', 'us',
             'use', 'via', 'way', 'went', 'without', 'would', 'arrived']

df_cloud = df_cloud.loc[~df_cloud['Word'].isin(stopwords), :]
df_cloud
```

```python
import warnings
warnings.filterwarnings('ignore')

sent_dropdown = alt.binding_radio(options=['Positive', 'Negative'], name='Sentiment: ')
sent_select = alt.selection_point(fields=['Pos-Neg'], bind=sent_dropdown, value='Positive')

single = alt.selection_single(on='mouseover')

base = alt.Chart(df_cloud).encode(
    x=alt.X('x:Q', scale=alt.Scale(domain=(-300, 3300))),
    y=alt.Y('y:Q', scale=alt.Scale(domain=(-300, 3300)))).add_params(
    sent_select
).transform_filter(
    sent_select
)

color = alt.Scale(domain=['Positive', 'Negative'], range=['green', 'red'])

word_cloud = base.mark_text(baseline='middle').encode(
    text='Word:N',
    size='Sum:Q',
    opacity=alt.condition(single, alt.value(1), 'datum.opacity', empty=False),
    color=alt.condition(single, alt.Color('Pos-Neg:N', scale=color), alt.value('lightgray')),
    tooltip=[alt.Tooltip('Word:N'), alt.Tooltip('Sum:Q', title='Count')]
).transform_calculate(
    x = 'random() * 3000',
    y = 'random() * 3000',
    opacity = '1 - (1 / datum.Sum) + 0.3'
).add_selection(
    single
).properties(
    title='Word Bubble based on Sentiment',
    height=400,
    width=400
).configure_title(
    font='Times New Roman',
    fontSize=15
).configure_axisX(
    disable=True
).configure_axisY(
    disable=True
).configure_legend(
    disable=True
).configure_view(
    stroke='black',
    strokeWidth=4,
    cornerRadius=5,
)

word_cloud.save('word_cloud.html')
```
